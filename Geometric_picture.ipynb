```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a grid of theta1 and theta2 values
theta1 = np.linspace(-5, 5, 100)
theta2 = np.linspace(-5, 5, 100)
Theta1, Theta2 = np.meshgrid(theta1, theta2)
```

```python
# Calculate the corresponding values of f(theta)
f_theta = Theta1**2 + Theta2**2
```

```python
# Create a contour plot
plt.contourf(Theta1, Theta2, f_theta, levels=20, cmap='viridis')
# Add labels and color bar
plt.xlabel('Theta1')
plt.ylabel('Theta2')
plt.colorbar(label='f(Theta1, Theta2)')
# Show the plot
plt.title('Contour Plot of f(Theta1, Theta2)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a grid of theta1 and theta2 values
theta1 = np.linspace(-5, 5, 100)
theta2 = np.linspace(-5, 5, 100)
Theta1, Theta2 = np.meshgrid(theta1, theta2)

# Values of K for level curves
K_values = [1, 2, 3, 4, 5]

# Create contour plots for different K values
for K in K_values:
    f_theta = Theta1**2 + Theta2**2 - K
    plt.contour(Theta1, Theta2, f_theta, levels=[0], colors='b', linestyles='solid')

# Add labels
plt.xlabel('Theta1')
plt.ylabel('Theta2')

# Show the plot
plt.title('Level Curves of f(Theta1, Theta2)')
plt.grid(True)
plt.axis('equal')  # Equal aspect ratio to make circles look like circles
plt.show()
```
