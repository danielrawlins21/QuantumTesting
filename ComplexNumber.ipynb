```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the complex number
z = 2 * np.exp(1j * np.pi/4)
```

```python
# Extract the real and imaginary parts
real_part = np.real(z)
imaginary_part = np.imag(z)
```

```python
print("Real representation ",real_part,"-  Imaginary representation",imaginary_part)
```

```python
# Create a plot
plt.figure(figsize=(6, 6))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-3, 3)
plt.ylim(-3, 3)

# Plot the complex number as a point
plt.scatter(real_part, imaginary_part, color='red', label='2 * e^(i*pi/4)')
plt.annotate('2 * e^(i*pi/4)', (real_part, imaginary_part), textcoords="offset points", xytext=(10, 10), ha='center')

# Add labels
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
